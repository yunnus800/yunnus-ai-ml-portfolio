```python
import pandas as pd
a=pd.read_csv("usl_dataset.csv")
print(a)
```

```python
object_cols_a = a.select_dtypes(include='object').columns.tolist()
print(object_cols_a)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
col=['FoodGroup', 'ShortDescrip', 'Descrip', 'CommonName', 'MfgName', 'ScientificName']
for i in col:
  a[i]=le.fit_transform(a[i])
```

```python
a.info()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
a= imputer.fit_transform(a)
print(a)
```

```python
import pandas as pd
df_a = pd.DataFrame(a)
df_a.duplicated().sum()
```

```python
a.shape
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

neighbors = NearestNeighbors(n_neighbors=90)
neighbors_fit = neighbors.fit(a)
distances, indices = neighbors_fit.kneighbors(a)
```

```python
import numpy as np
distances = np.sort(distances[:, 44], axis=0)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(distances)
plt.title("K-Distance Graph (To find Epsilon)")
plt.xlabel("Data Points sorted by distance")
plt.ylabel("Epsilon (Distance)")
plt.grid(True)
plt.show()
```

```python
dbscan = DBSCAN(eps=400, min_samples=90)
clusters = dbscan.fit_predict(a)
```

```python
if len(set(clusters)) > 1:
    score = silhouette_score(a, clusters)
    print(f"Silhouette Score: {score:.3f}")
else:
    print("DBSCAN found only one cluster or all noise. Adjust eps/min_samples.")
```

```python
print("Cluster Labels (including -1 for noise):", np.unique(clusters))
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.scatter(
    a[:, 0],   # PC1
    a[:, 44],   # PC2
    c= clusters,           # cluster labels (DBSCAN output)
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("DBSCAN Clusters")

plt.show()
```

```python
plt.scatter(x=a[:,0],y=a[:,44],c=clusters)
```

```python
import pandas as pd
val = pd.Series(clusters).value_counts()
print(val)
```

```python
from sklearn.preprocessing import StandardScaler
a_scaled = StandardScaler().fit_transform(df_a)
print(a_scaled)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state = 42)
kmeans.fit(a_scaled)
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(a_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=2,init="k-means++")
kmeans.fit(a_scaled)
```

```python
print(kmeans.labels_)
```

```python
print(kmeans.cluster_centers_)
```

```python
from sklearn.metrics import silhouette_score

silhouette_coefficients = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(a_scaled)
    score = silhouette_score(a_scaled, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```
