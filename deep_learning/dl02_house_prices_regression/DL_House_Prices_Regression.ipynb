```python
import pandas as pd
from sklearn.model_selection import train_test_split


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.initializers import HeNormal
```

```python
data = pd.read_csv('house_prices.csv')
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
object_cols = data.select_dtypes(include='object').columns.tolist()
print(object_cols)
```

```python
data[object_cols] = data[object_cols].fillna('Missing')

label_encoders = {}

for col in object_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
x = data.drop('SalePrice', axis=1)
y = data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
model = Sequential()
model.add(Dense(units=64, activation='relu', kernel_initializer=HeNormal()))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='linear'))

# Compile
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mae'])
```

```python
model.fit(xtrain, ytrain, epochs=50, batch_size=32, verbose=0)
```

```python
loss, mae = model.evaluate(xtest, ytest, verbose=0)
print(f"Test MAE: {mae:.2f}")
```
